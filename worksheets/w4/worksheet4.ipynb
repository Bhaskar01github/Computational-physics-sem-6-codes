```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler(f1,x,y,h):
    return y + h*f1(x,y)

def midpoint(f1,x,y,h):
    k1=h*f1(x,y)
    return y + h*f1(x+h/2, y+k1/2)

def call_method(method,fn,y_i,h):
    y=y_i
    ys=[y]
    t=np.arange(0,1,h)
    for i in range(0,len(t)-1):
        x=t[i]
        y=method(fn,x,y,h)
        ys.append(y)
    a=t[-1]    
    return (ys,a)   
    
```
